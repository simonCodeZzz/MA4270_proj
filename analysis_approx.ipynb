# This notebook compares the true kernel and Random Fourier Features approximation in terms of:
## 1. Time taken to transform the data
## 2. MSE between the kernel matrices

```python
import numpy as np
import matplotlib as mpl
%matplotlib inline 
from matplotlib import pyplot as plt
```

## Simulated Data

```python
Ds = np.load("data/Ds.npy")
rbf_mse = np.load("data/rbf_mse.npy")
laplace_mse = np.load("data/laplace_mse.npy")
rbf_time = np.load("data/rbf_time.npy")
laplace_time = np.load("data/laplace_time.npy")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,4))
for ax,data,title in zip(axes,[rbf_mse, laplace_mse],['RBF Kernel','Laplacian Kernel']):
    ax.plot(Ds, data)
    ax.set_ylabel("MSE (log)")
    ax.set_xlabel("Number of MC samples D")
    ax.set_yscale("log")
    ax.set_title(title)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,4))
for ax,data,title in zip(axes,[rbf_time, laplace_time],['RBF Kernel','Laplacian Kernel']):
    ax.plot(Ds, data[1:])
    ax.hlines(y=data[0], xmin=min(Ds), xmax=max(Ds),colors="red")
    ax.set_ylabel("Time (sec)")
    ax.set_xlabel("Number of MC samples D")
    ax.set_title(title)
plt.show()
```

## Fashion MNIST (classification)

```python
Ds = np.load("data/Ds_fashion.npy")
rbf_mse = np.load("data/rbf_mse_fashion.npy")
laplace_mse = np.load("data/laplace_mse_fashion.npy")
rbf_time = np.load("data/rbf_time_fashion.npy")
laplace_time = np.load("data/laplace_time_fashion.npy")
logDs = np.log2(Ds)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,4))
for ax,data,title in zip(axes,[rbf_mse, laplace_mse],['RBF Kernel','Laplacian Kernel']):
    ax.plot(logDs, data)
    ax.set_ylabel("MSE (log)")
    ax.set_xlabel("Number of MC samples D (log2)")
    ax.set_yscale("log")
    ax.set_title(title)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,4))
for ax,data,title in zip(axes,[rbf_time, laplace_time],['RBF Kernel','Laplacian Kernel']):
    ax.plot(logDs, data[1:],label='Approximation')
    ax.hlines(y=data[0], xmin=min(logDs), xmax=max(logDs),colors="red", label='True Kernel')
    ax.set_ylabel("Time (sec)")
    ax.set_xlabel("Number of MC samples D (log2)")
    ax.set_title(title)

plt.legend(bbox_to_anchor=[0,1.2])
plt.show()
```
