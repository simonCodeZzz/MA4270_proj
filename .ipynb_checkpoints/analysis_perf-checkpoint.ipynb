# This notebook compares the true kernel and Random Fourier Features approximation in terms of:
## 1. Time taken to fit and train data
## 2. Test Performance (Error Rate & RMSE)

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
%matplotlib inline 
from matplotlib import pyplot as plt
```

## Fashion MNIST (classification)

```python
exact_errors = np.load("data/exact_errors_fashion.npy")
exact_time = np.load("data/exact_time_fashion.npy")
rff_errors = np.load("data/rff_errors_fashion.npy")
rff_times = np.load('data/rff_times_fashion.npy')
Ds = np.load('data/Ds_fashion.npy')
n_data = np.load('data/data_num_fashion.npy')
```

```python
f = plt.figure(figsize=(24,5))

plt.subplot(121)
for i, line in enumerate(Ds):
    plt.plot(n_data, rff_rmse[i,:], label=Ds[i], linestyle='dashed', alpha=0.6)
plt.plot(n_data,exact_rmse, label="Kernel Ridge", color='black')
plt.ylabel("Testing Error")
plt.xlabel("Number of data")

plt.subplot(122)
for i, line in enumerate(Ds):
    plt.plot(n_data, rff_times[i,:], label=Ds[i], linestyle='dashed', alpha=0.6)
plt.plot(n_data,exact_time, label="Kernel Ridge", color='black')
plt.ylabel("Training Time(s)")
plt.xlabel("Number of data")
plt.legend(bbox_to_anchor=[1,0.6])
plt.tight_layout()
```

## California Housing (Regression)

```python
exact_rmse = np.load("data/exact_rmse_housing.npy")
exact_time = np.load("data/exact_time_housing.npy")
rff_rmse = np.load("data/rff_rmse_housing.npy")
rff_times = np.load('data/rff_times_housing.npy')
Ds = np.load('data/Ds_housing.npy')
n_data = np.load('data/data_num_housing.npy')
```

```python
f = plt.figure(figsize=(24,5))

plt.subplot(121)
for i, line in enumerate(Ds):
    plt.plot(n_data, rff_rmse[i,:], label=Ds[i], linestyle='dashed', alpha=0.6)
plt.plot(n_data,exact_rmse, label="Kernel Ridge", color='black')
plt.ylabel("Testing RMSE")
plt.xlabel("Number of data")

plt.subplot(122)
for i, line in enumerate(Ds):
    plt.plot(n_data, rff_times[i,:], label=Ds[i], linestyle='dashed', alpha=0.6)
plt.plot(n_data,exact_time, label="Kernel Ridge", color='black')
plt.ylabel("Training Time(s)")
plt.xlabel("Number of data")
plt.legend(bbox_to_anchor=[1,0.6])
plt.tight_layout()
plt.savefig("figs/housing_rmse.png")
```
